```python
import seaborn as sns
corr = medical_report.corr()
ax, fig = plt.subplots(figsize=(15,15))
sns.heatmap(corr, vmin=-1, cmap='coolwarm', annot=True)
plt.show()
```

```python
#The features 'age', 'ejection_fraction', 'serum_creatinine', 'serum_sodium', and 'time' have a considerable correlation with 'DEATH_EVENT'
X=medical_report[['age','ejection_fraction','serum_creatinine','serum_sodium','time']]
Y=medical_report['DEATH_EVENT']
```

```python
from sklearn.preprocessing import StandardScaler
# define standard scaler
scaler = StandardScaler()
# transform data
scaled_X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(scaled_X,Y,test_size = 0.2,random_state=2)
```

```python
#creating svm classifier
from sklearn.svm import SVC
model = SVC(kernel='linear',C=1.5,gamma="auto")
```

```python
model.fit(X_train, Y_train)
```

```python
model.score(X_train, Y_train)
```

```python
model.get_params()
```

```python
predict = model.predict(X_test)
```

```python
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",(metrics.accuracy_score(Y_test, predict)*100))
```

```python
print("Error: ",(1-metrics.accuracy_score(Y_test, predict))*100)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test, predict))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test, predict) 
```

```python
confusion=confusion_matrix(Y_test, predict) 
ax, fig = plt.subplots()
sns.heatmap(confusion, vmin=-1, cmap='coolwarm', annot=True)
plt.show()
```
