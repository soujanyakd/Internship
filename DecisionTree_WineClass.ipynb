```python
wine_data.drop('Ash',inplace=True, axis=1)
```

```python
features=wine_data.drop('Class',axis=1)
target=wine_data['Class']
```

```python
from sklearn.preprocessing import StandardScaler
# define standard scaler
scaler = StandardScaler()
# transform data
scaled_features = scaler.fit_transform(features)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(scaled_features,target,test_size = 0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion='gini',max_depth=10,min_samples_leaf=10)
model.fit(X_train,Y_train)
```

```python
model.score(X_train, Y_train)
```

```python
predict = model.predict(X_test)
```

```python
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",(metrics.accuracy_score(Y_test, predict)*100))
```

```python
print("Error: ",(1-metrics.accuracy_score(Y_test, predict))*100)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test, predict))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test, predict) 
```

```python
confusion=confusion_matrix(Y_test, predict) 
ax, fig = plt.subplots()
sns.heatmap(confusion,cmap='RdBu',annot=True)
plt.show()
```

```python
confusion.shape
```

```python
confusion=confusion_matrix(Y_test, predict)
actual=[]
predicted=[]
for i in range(3):
    for j in range(3):
        actual.append(confusion[i][j])
        if i!=j:
            predicted.append(confusion[i][j])
print(actual)
print(predicted)
```

```python
error=(np.sum(predicted)/np.sum(actual))*100
print(error)
```
